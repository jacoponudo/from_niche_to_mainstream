```python
from EXP_package.read_data import *
from EXP_package.entropy import *
from EXP_package.functions import *
import os
platforms = ['facebook', 'reddit', 'gab', 'voat']
save_folder = '/home/jacoponudo/Documents/Size_effects/PLT/5_dynamic_entropy'
max_size = 1000
step = 20

# Ensure the save folder exists
os.makedirs(save_folder, exist_ok=True)

for platform in platforms:
    df = read_data(platform)
    df = process_posts(df)
    entropy_by_class = calculate_entropy_by_class(df, max_size=max_size, step=step)
    
    # Save the plot
    plot_entropy(entropy_by_class, platform)
    save_path = os.path.join(save_folder, f'{platform}_dynamic_entropy.png')
    
    # Assuming plot_entropy function saves the plot, alternatively you might need to use plt.savefig()
    plt.savefig(save_path)
    plt.close()
```
